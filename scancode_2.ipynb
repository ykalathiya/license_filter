```python
import os
import json
from scancode import api

def scan_files_in_directory(directory, output_file):
    all_scan_results = []

    # Walk through the directory structure
    for root, dirs, files in os.walk(directory):
        for file in files:
            if '.git' in dirs:
                dirs.remove('.git')

            file_path = os.path.join(root, file)
            print(f"Scanning file: {file_path}")
            
            try:
                # Run the scan for the current file
                scan_license = api.get_licenses(file_path, unknown_licenses=True)
                scan_copyright = api.get_copyrights(file_path)
                
                # Store the results along with the file path
                all_scan_results.append({
                    'file_path': file,
                    'license': scan_license,
                    'copyright' : scan_copyright
                })
            except Exception as e:
                print(f"An error occurred while scanning {file_path}: {e}")
    
    # Save all results to the output JSON file
    with open(output_file, 'w') as f:
        json.dump(all_scan_results, f, indent=2)
    
    print(f'Scan completed. Results saved to {output_file}')

# Define the path to the root directory you want to scan
root_directory = 'D:\\code\\license_remove\\hello-world'

# Define the path for the output JSON file
output_file = 'output_file.json'

# Run the scan
scan_files_in_directory(root_directory, output_file)
```

```python
import subprocess
def scanCode(output_file, code_dir):
    try:
        # Run the ScanCode-Toolkit command
        result = subprocess.run(
            ['scancode', '-cl', '--json-pp', output_file, code_dir],
            check=True,
            capture_output=True,
            text=True
        )
        print(f"ScanCode-Toolkit output:\n{result}")
        if result.stderr:
            print(f"ScanCode-Toolkit errors:\n{result.stderr}")
    except FileNotFoundError as e:
        print("Error: File not found. ",e)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

**Fetch index from JSON to dictionary**

```python
import json

def fetchIndex(json_file_path):
    # Load JSON data from the file
    with open(json_file_path, 'r') as file:
        data = json.load(file)

    license_index = {}

    # Iterate through each file in the JSON data
    for file in data.get('files', []):
        file_path = file.get('path')
        
        # Check for license detections in the file
        for license_detection in file.get('license_detections', []):
            for match in license_detection.get('matches', []):
                start_line = match.get('start_line')
                end_line = match.get('end_line')
                
                if file_path and start_line is not None and end_line is not None:
                    print(f"{file_path} : {start_line} to {end_line}")
                    if file_path not in license_index:
                        license_index[file_path] = []
                    license_index[file_path].extend(range(start_line, end_line + 1))
        
        # Check for copyright information in the file
        for copyright in file.get('copyrights', []):
            start_line = copyright.get('start_line')
            end_line = copyright.get('end_line')
            
            if file_path and start_line is not None and end_line is not None:
                print(f"{file_path} : {start_line} to {end_line}")
                if file_path not in license_index:
                    license_index[file_path] = []
                license_index[file_path].extend(range(start_line, end_line + 1))

    # Remove duplicate line numbers
    for file_path in license_index:
        license_index[file_path] = sorted(set(license_index[file_path]))

    return license_index
```

```python
import os

def count_files_in_directory(directory):
    """
    Count the number of files in a directory and its subdirectories.
    Print each file's name.
    """
    total_files = 0
    for root, dirs, files in os.walk(directory):
        if '.git' in dirs:
            dirs.remove('.git')
        for file in files:
            total_files += 1
            print(f"File: {os.path.join(root, file)}")
    return total_files

def main():
    directory = 'D:\\code\\license_remove\\openj9'  # Replace with your directory path
    total_files = count_files_in_directory(directory)
    print(f"Total number of files in '{directory}': {total_files}")

if __name__ == '__main__':
    main()
```
